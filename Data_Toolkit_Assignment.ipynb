## 2) The difference between Deep copy and shallow copy.

# Deep Copy :- A deep Copy creates a completely independent copy of both the array and its data. It does not share any data with the original array.

- A deep copy means actually creating a new array and copying over the values and the new variable and existing variable will point to the different memory location,

  So if there is any change in new variable or existing variable it will no reflect to the other variable data.

- Syntax used in deep copy :------>            ***.copy()***



```python
# Example of Deep Copy :-
# Create a 1D array

org_arr = np.array([1, 2, 3, 4, 5, 6])
org_arr
```

```python
# Create deep copy array from orignal array :-

deep_copy_arr = org_arr.copy()
deep_copy_arr
```

```python
# If you make change is any individual array it will not refelect to the other array :-
# Change the value of deep_copy_arr from 2 to 200
# Used index number to access the element of the list

deep_copy_arr[1] = 200
deep_copy_arr
```

```python
# Check the original arry it will be the same without any change :-

org_arr
```

# Shallow Copy :- A shallow copy creates a new array, but it does not create new copies of the elements within the array. Instead, it points to the same elements as the original array

- A shallow copy refers to specific values that are still connected to the original Variable (Array or Object). Every change to a new variable will affect the original variable, and the new variable and existing variable will point to the same memory location,

  So if there is any change in new variable or existing variable it will reflect on both.

- Syntax used in shallow copy :------>            *** = ***

```python
# Example of Shallow Copy :-
# Create a 1D array

a = np.array([7, 8, 9, 10, 11, 12])
a
```

```python
# Create shallow copy array from orignal array :-

shallow_copy_arr = a
shallow_copy_arr
```

```python
# If you make change is any individual array it will refelect to the same change in other array also :-
# Change the value of shallow_copy_arr from 11 to 50000
# Used index number to access the element of the list

shallow_copy_arr[-2] = 50000
shallow_copy_arr
```

```python
# Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity'
# Let's assume ------> assume random date values e.g., from 1-10-2023, periods/size = 100 and frequency = M(monthly wise)
# assume random temperature values e.g., between -10 and 35 degrees Celsius and size = 100
# assume random humidity values e.g., between 0 and 100 % and size = 100


date = pd.date_range(start = "2023-07-10", periods = 100, freq = "M")
temp = np.random.uniform(low = -10, high = 35, size = 100)
humidity = np.random.uniform(low = 0, high = 100, size = 100)

time_series_df = pd.DataFrame({"Date" : date, "Temperature" : temp, "Humidity" : humidity})
time_series_df
```

```python
# a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and right y-axis for 'Humidity').

# 1st step :- create a figure/page for the plotting :- # "fig, ax1 = plt.subplots()"
# 2nd step ------> create 1st plot with the data details, color, label for temperature with date
# 3rd step ------> mention y-axis label "temperature"

# 4th step ------> create a twin y - axes sharing the xaxis
# To create a twin y - axes sharing the xaxis :- use this code
# ax1 = plt.gca()
# ax2 = ax1.twinx()

# 5th step ------> create 2nd plot with the data details, color, label for humidity with date
# 6th step ------> mention y-axis label "humidity"
# 7th step ------> apply plt.legend() to show the label
# 8th step ------> apply plt.show() to show the plot




# create a figure/page for the plotting with the plot size :-
fig, ax1 = plt.subplots(figsize = (10, 6))


# create 1st plot of temperature with the data details, color, label on the left y-axis with date :-
ax1.plot(time_series_df["Date"], time_series_df["Temperature"], color="blue", label="Temperature")
ax1.set_ylabel("Temperature", color="blue")

# create a twin axis for Humidity, sharing on the x-axis "date" with temperature y-axis :-
ax2 = ax1.twinx()

# create 2nd plot of humidity with the data details, color, label on the left y-axis with date :-
ax2.plot(time_series_df["Date"], time_series_df["Humidity"], color="red", label="Humidity")
ax2.set_ylabel("Humidity", color="red")

fig.legend()
plt.show()
```

```python
# b) Label the x-axis as 'Date'.


fig, ax1 = plt.subplots(figsize = (10, 6))


ax1.plot(time_series_df["Date"], time_series_df["Temperature"], color="blue", label="Temperature")
ax1.set_ylabel("Temperature", color="blue")

ax2 = ax1.twinx()

ax2.plot(time_series_df["Date"], time_series_df["Humidity"], color="red", label="Humidity")
ax2.set_ylabel("Humidity", color="red")

# created Label the x-axis as 'Date'.
ax1.set_xlabel("Date")

fig.legend()
plt.show()
```

```python
# c) Set the title of the plot as 'Temperature and Humidity Over Time'.


fig, ax1 = plt.subplots(figsize = (10, 6))

ax1.plot(time_series_df["Date"], time_series_df["Temperature"], color="blue", label="Temperature")
ax1.set_ylabel("Temperature", color="blue")

ax2 = ax1.twinx()

ax2.plot(time_series_df["Date"], time_series_df["Humidity"], color="red", label="Humidity")
ax2.set_ylabel("Humidity", color="red")

ax1.set_xlabel("Date")

# mentioned the title of the plot as 'Temperature and Humidity Over Time' :-
plt.title("Temperature and Humidity Over Time")

fig.legend()
plt.show()
```

## Answer 19 :- Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_file

from bokeh.models import ColumnDataSource, HoverTool

from bokeh.palettes import Viridis256

from bokeh.transform import linear_cmap
```

```python
# generated categorical data

classses = ["Class 6th", "Class 7th", "Class 8th", "Class 9th", "Class 10th", "Class 11th", "Class 12th"]

students = [19, 50, 37, 55, 29, 69, 49]

Class_Room_Data = pd.DataFrame({"Classes" : classses, "Students" : students})

Class_Room_Data
```

```python
# Generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'.

# 1st ---> Created a source for ColumnDataSource of dataframe -------> source = ColumnDataSource(Class_Room_Data)
# 2nd ---> Created a figure for the plot with title name and x_range of classes
# 3rd ---> Created a color_map to define a color mapper to color bars based on their values
# 4th ---> Used bokeh vbar(vertical bar) function to make plot providing all the necessary details like (x and y column name, color detailing, bar width size)
# 5th ---> Add hover tooltips to display exact value
# 6th ---> Provide x and y axis labels to the chart
# 7th ---> Used show(p) to show the chart


source = ColumnDataSource(Class_Room_Data)

p = figure(x_range = Class_Room_Data['Classes'], title = "Random Categorical Bar Chart")

color_map = linear_cmap(field_name = 'Students', palette = Viridis256, low = min(students), high = max(students))

p.vbar(x = 'Classes', top = 'Students', source = source, line_color = 'black', fill_color = color_map)


hover = HoverTool()
hover.tooltips = [("Class", "@Classes"), ("Number of Students", "@Students")]
p.add_tools(hover)


p.xaxis.axis_label = "Class"
p.yaxis.axis_label = "Number of Students'"

show(p)
```

## Answer 20 :- Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as 'Simple Line Plot'

```python
import plotly.graph_objects as go

import plotly.express as px
```

#### There are two ways to create basic line plot

1st way to create line plot from plotly graph_object (go)

2nd way to create line plot from plotly express (px)

```python
# randomly generated dataset using numpy :-
# store this into dataframe :-

x = np.arange(50)
y = np.random.randn(50)

random_df = pd.DataFrame({"X" : x, "Y" : y})
random_df
```

#### 1st way to create line plot from plotly graph_object (go) :-

```python
# Created a basic line plot of dataframe "random_df", label the axes, and set the title as 'Simple Line Plot'
# Steps :-
# 1st ----> Applied figure finction to give figure to the plot
# 2nd ----> Used plotly graph_object() code to provide details to the chart and to show line chart
# 3rd ----> Applied figure update layout to show label the axes, and set the title as 'Simple Line Plot'
# 4th ----> Used show() to show the plot.

fig = go.Figure()

fig.add_trace(go.Scatter(x = random_df.X, y = random_df.Y, mode = "lines"))

fig.update_layout(title = "Simple Line Plot", xaxis_title = "X Axis", yaxis_title = "Y Axis")


fig.show()
```

#### 2nd way to create line plot from plotly express (px) :-

```python
# Created a basic line plot of dataframe "random_df", label the axes, and set the title as 'Simple Line Plot'
# Steps :-
# 1st ---->
# 2nd ----> Used plotly express() code to provide details to the chart and and set the title as 'Simple Line Plot'
# 3rd ----> Applied figure update layout to show label the axes,
# 4th ----> Used show() to show the plot.


fig = px.line(random_df, x = "X", y = "Y", title = "Simple Line Plot", markers = True)

fig.update_xaxes(title_text = "X Axis")
fig.update_yaxes(title_text = "Y Axis")

fig.show()
```

## Answer 21 :- Using Plotly, created an interactive pie chart of randomly generated data, add labels and percentages, set the title as 'Interactive Pie Chart'.

### Created pie chart in two ways :-

1st way to create pie chat using plotly graph_object (go) :-

2nd way to create pie chat using plotly express (px) :-


```python
# Generated a data of subjects and generate random marks
# Store this into dataframe

subjects = ["Maths", "English", "Arts", "Commerce", "Science", "Hindi"]
marks = [73, 87, 59, 93, 75, 69]

subject_marks_df = pd.DataFrame({"Subjects" : subjects, "Marks" : marks})
subject_marks_df
```

#### 1st way to create pie chat using plotly graph_object (go) :-

```python
# Create an interactive pie chart, add labels and percentages, set the title as 'Interactive Pie Chart'.

# Using plotly graph_object (go) :-

# Steps :-
# 1st ----> Used plotly graph_object(go) ----> fig = go.Figure() with the required data :-
# go.Pie() ---> is used to create the pie chart.
# labels ---> specifies the slice labels.
# values ---> specifies the slice sizes.
# textinfo = "label+percent" ---> to shows the label and percentage on the pie slices.
# pull ---> to give "explode" effect to the pie chart

# 2nd ----> Used (fig.update_layout()) ---> to provide title to the pie chart

# 3rd ----> applied fig.show() to display the chart


fig = go.Figure([go.Pie(labels = subject_marks_df["Subjects"], values = subject_marks_df["Marks"], textinfo = "label + percent", pull = [0.0, 0.1, 0.2, 0.3, 0.1, 0.2])])


fig.update_layout(title = "Interactive Pie Chart")


fig.show()
```

#### 2nd way to create pie chat using plotly express (px) :-

```python
# Create an interactive pie chart, add labels and percentages, set the title as 'Interactive Pie Chart'.

# Using plotly express (px) :-

# Steps :-
# 1st ----> Used plotly plotly express (px) ----> fig = px.pie() with the required data :-
# datafarme name ---> provided the dataframe variable name to create the pie chart.
# names ---> specifies the slice labels.
# values ---> specifies the slice sizes.
# title ---> to show the title name to the pie chart.
# label ---> provided subjects and marks name to show on the pie chart
# hole = value ---> to show hole in the mid of pie chart

# 2nd applied fig.show() to display the chart


fig = px.pie(subject_marks_df, names = "Subjects", values = "Marks", title = "Interactive Pie Chart", labels = {'Subjects': 'Subject', 'Marks': 'Marks'}, hole = 0.2)

fig.show()
```
